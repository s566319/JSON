2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
#Read in the contents of your file. Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics. Print the polarity score of the sentiment analysis. Given that the range of the polarity score is [-1.0,1.0] which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion? Answer this question in a comment in your code cell.
import requests
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load spaCy model
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

# Read the data
with open('lyrics.json', 'r') as json_file:
    results = json.load(json_file)

# Check if the JSON data has 'lyrics'
if 'lyrics' in results:
    # Pull out the lyrics
    song = results['lyrics']
    
    # Process the lyrics using spaCy
    doc = nlp(song)

    # Perform sentiment analysis
    sentiment_score = doc._.polarity

    # Print the lyrics
    print("Lyrics of the song:")
    print(song)

    # Print the polarity score of the sentiment analysis
    print("Polarity score:", sentiment_score)
    # Determine sentiment
    if sentiment_score > 0:
        print("The lyrics have more of a positive connotation.")
    elif sentiment_score < 0:
        print("The lyrics have more of a negative connotation.")
    else:
        print("The lyrics are neutral.")
else:
    print("There are no lyrics")

```

```python
## The polarity score ranges from -1 to 1.  A score of -1 means the words are negative, and a score of 1 means the words are super positive.  Since the spaCyTextBlob determined the score is 0.044, I would say the lyrics are neutural, but little more positive, rather than negative 
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

def write_lyrics_to_json(artist, song, filename):
    # Build the API URL
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"

    # Send a GET request to the API
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Extract the JSON data from the response
        data = response.json()

        # Check if the lyrics are there
        if 'lyrics' in data:
            # Write the JSON data to the specified file with (w for write mode)
            with open(filename, 'w') as file:
            # The file is more readable
                json.dump(data, file, indent=4)
# print lyrics
            print(f"Lyrics for '{song}' by '{artist}' have been written to '{filename}'.")
        else:
            print(f"Can not find lyrics to '{song}' by '{artist}'.")
    else:
        print(f"Failed to find lyrics for '{song}' by '{artist}'.")

# Writting files
write_lyrics_to_json("Morgan Wallen", "Last Night", "night_morgan_lyrics.json")
write_lyrics_to_json("Taylor Swift", "Anti-Hero", "taylor_antihero_lyrics.json")
write_lyrics_to_json("Drake", "One Dance", "drake_dance_lyrics.json")
write_lyrics_to_json("Roxette", "The Look", "roxette_look_lyrics.json")
```
